## Feature Engineering

```python
import numpy as np

# Creating new features
data['Income per Age'] = data['Annual Income (k$)'] / data['Age']
data['Spending per Age'] = data['Spending Score (1-100)'] / data['Age']

# Handle any potential infinities from division by zero (if age is zero, which is unlikely but good practice)
data.replace([np.inf, -np.inf], np.nan, inplace=True)
data.fillna(0, inplace=True)  # Assuming replacing with zero is reasonable here

# Verify new features
print(data[['Income per Age', 'Spending per Age']].head())
```

## Preparing Data for Clustering

```python
# Feature selection including new features
selected_features = data[['Annual Income (k$)', 'Spending Score (1-100)', 'Income per Age', 'Spending per Age']]

# Standardize selected features
scaler = StandardScaler()
features_scaled_new = scaler.fit_transform(selected_features)

# Verify scaling
print(pd.DataFrame(features_scaled_new, columns=selected_features.columns).head())
```

```python
# Dimensionality Reduction for Visualization

from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score

pca = PCA(n_components=2)
pca_result = pca.fit_transform(features_scaled_new)
```

## Advanced Clustering Techniques

### K-Means Clustering

```python
from sklearn.cluster import KMeans

# Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(features_scaled_new)
    wcss.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Fit K-Means model with the optimal number of clusters
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
data['KMeans_Cluster'] = kmeans.fit_predict(features_scaled_new)
```

```python
data['KMeans_Cluster']
```

```python
# Clustering Evaluation
silhouette_avg = silhouette_score(features_scaled_new, data['KMeans_Cluster'])
davies_bouldin = davies_bouldin_score(features_scaled_new, data['KMeans_Cluster'])
print("Silhouette Score:", silhouette_avg)
print("Davies-Bouldin Score:", davies_bouldin)
```

```python
# Cluster centers
kmeans_centers = kmeans.cluster_centers_
print(kmeans_centers)

# Analyze each K-Means cluster
for i in range(3):
    cluster_data = data[data['KMeans_Cluster'] == i]
    print(f'K-Means Cluster {i} Summary:')
    print(cluster_data.describe())
```

```python
# Visualize K-Means clusters
plt.figure(figsize=(10, 5))
sns.scatterplot(data=data, x='Annual Income (k$)', y='Spending Score (1-100)', hue='KMeans_Cluster', palette='viridis')
plt.title('K-Means Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Visualize Hierarchical clusters
plt.figure(figsize=(10, 5))
sns.scatterplot(data=data, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Hierarchical_Cluster', palette='viridis')
plt.title('Hierarchical Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Visualizing Clusters using PCA
plt.figure(figsize=(10, 7))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=data['KMeans_Cluster'], palette='viridis', s=100)
plt.title('Customer Segments based on Purchasing Behavior and Demographics (PCA)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='K_means')
plt.grid(True)
plt.show()
```

```python
# Analyze each segment
# Define the features you used for clustering
features = ['Annual Income (k$)', 'Spending Score (1-100)', 'Income per Age', 'Spending per Age']

# Group data by segment
segment_means = data.groupby('KMeans_Cluster')[features].mean()
segment_counts = data['KMeans_Cluster'].value_counts().sort_index()

# Print segment analysis
print("\nSegment Analysis:")
for segment, count in segment_counts.items():
    print(f"\nSegment {segment} - Count: {count}")
    print(segment_means.loc[segment])
```
